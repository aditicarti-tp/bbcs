```python
# Reorder the columns
df = df[['iso3c', 'Country_Name', 'trend', 'forests_2000', 'forests_2020']]
df
df=df.to_csv("cleaneddeforestation.csv",index=False)
```

```python
#make everything easier to read by removing decimels
df=pd.read_csv('cleaneddeforestation.csv')
df['trend'] = df['trend'].round(0)
df['forests_2000'] = df['forests_2000'].round(0)
df['forests_2020'] = df['forests_2020'].round(0)
df
df['trend'] = df['trend'].astype(int)
df['forests_2000'] = df['forests_2000'].astype(int)
df['forests_2020'] = df['forests_2020'].astype(int)

df
```

```python
#make index start from 1
df.reset_index(drop=True, inplace=True)  # Reset index, dropping the old one
df.index = df.index + 1  # Shift index to start from 1
df
```

```python
#remove all the new forests_2000 and forests_2020 = 0
df = df[~((df["forests_2000"] == 0.0) & (~(df["forests_2020"] == 0.0)))]
df
df=df.to_csv('cleaned_deforestation.csv')
```
